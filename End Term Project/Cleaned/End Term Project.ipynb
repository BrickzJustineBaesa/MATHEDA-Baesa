```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"..\raw\data_science_job.csv",
                 delimiter=",")
df.info()
```

```python
df_nn = df.rename(columns={
'enrollee_id':'Student_ID',
'city':'City No.',
'city_development_index':'CityDev. Index',
'gender':'Sex',
'relevent_experience':'Relevent Exp.',
'enrolled_university':'Enrolled Uni.',
'education_level':'Ed. Level',
'major_discipline':'Major Discipline',
'experience':'Exp.',
'company_size':'Comp. Size',
'company_type':'Comp. Type',
'training_hours':'Training Hrs',
'target':'Target'
})
df_nn.info()
```

```python
df_nn = df_nn.fillna('Unknown')
df_nn
```

```python
df_cnt=df_nn['City No.'].value_counts()
df_cnt= df_cnt.sort_values(ascending=False)
df_cnt.head(10)
```

```python
chart = df_cnt.head(10).plot(kind='bar',
                color =['#C4A484'] 
                ,figsize=(10,5))
chart.bar_label(chart.containers[0],label_type='edge', 
                fontfamily = 'Times New Roman',
                fontweight = 'bold')
plt.title('Top 10 Cities with the Highest No. of Applicants',
          fontfamily = 'Times New Roman',
          fontweight = 'bold',
          fontsize='18')
plt.xlabel('City No.',
           
        fontfamily = 'Times New Roman',
          fontweight = 'bold',
          fontsize='16')
plt.ylabel('No. of Students',
           fontfamily = 'Times New Roman',
          fontweight = 'bold',
          fontsize='16')
plt.xticks(rotation=0, ha='center')
plt.show()
```
